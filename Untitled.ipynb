```python
# define GALEX bandwidth
GALEX_units =  1*u.photon / u.cm**2 / u.s / u.Angstrom
wavel_min = 1350 * u.Angstrom
wavel_max = 1750 * u.Angstrom
print("Galex data is", wavel_min.value, "-", wavel_max)
f_max_hz = (cst.c.cgs / wavel_min.to(u.cm)).to(u.Hz)
f_min_hz = (cst.c.cgs / wavel_max.to(u.cm)).to(u.Hz)
print("       ","{:e}".format(f_min_hz.to(u.GHz).value), "-", "{:e}".format(f_max_hz.to(u.GHz)))

nu = (f_max_hz + f_min_hz) / 2
```

```python
# conversion array from erg/s/Hz/cm2 to photons/s/A/cm2
def convert_to_galex_units(F_erg_hz_cm2,nu):
    erg_hz_cm2 = 1*u.erg/u.s/u.Hz/u.cm**2
    w = nu.to(u.AA, equivalencies=u.spectral())
    C = (erg_hz_cm2).to(GALEX_units, u.spectral_density(w)) / erg_hz_cm2 * 2*np.pi
    #                                                                      ^^^^^^^ this comes from using hbar instead of h in the conversion.
    return F_erg_hz_cm2 * C
```

```python
# calculate AQN temperature
T_aqn_ev = 5 * u.eV # this is an estimate for the warm gas case
T_aqn_ev = 0.12 * u.eV
```

```python
# erg Hz^-1 s^-1 cm^-2
# nu Hz
# T eV
def spectral_surface_emissivity(nu, T):
    T = T * eV_to_erg
    w = 2 * np.pi * nu * Hz_to_erg
    unit_factor = (1 / cst.hbar.cgs) * (1/(cst.hbar.cgs * cst.c.cgs))**2 * (cst.hbar.cgs * 1/u.Hz * 1/u.s)
    #                ^ 1/seconds           ^ 1/area                          ^ 1/frequency and energy
    return unit_factor * 4/45 * T**3 * cst.alpha ** (5/2) * 1/np.pi * (T/(m_e_eV*eV_to_erg))**(1/4) * (1 + w/T) * np.exp(- w/T) * h(w/T)

# erg Hz^-1 s^-1 cm^-2
# nu Hz
# T eV
def spectral_surface_emissivity_cgs(nu, T):
    T_K = T * eV_to_K
    k_B_cgs = cst.k_B.cgs
    h_cgs = cst.h.cgs

    return  (1/u.Hz)*(1/cst.hbar.cgs * 1/cst.c.cgs)**2 * 4/45 * (k_B_cgs * T_K)**3 * cst.alpha ** (5/2) * 1/np.pi * ((k_B_cgs * T_K)/(m_e_eV*eV_to_erg))**(1/4) * (1 + (h_cgs * nu)/(k_B_cgs * T_K)) * np.exp(- (h_cgs * nu)/(k_B_cgs * T_K)) * h((h_cgs * nu)/(k_B_cgs * T_K))/u.s

spectral_surface_emissivity_cgs(nu, T_aqn_ev)
```
